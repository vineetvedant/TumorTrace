# here is the main code which take video as input and give the output 
 # it take mp4 and give the result in same formate 

```python
import os
from ultralytics import YOLO
import cv2

# Define the input video file path (Your new input video path)
video_path = r'C:\Users\VEDANT\OneDrive\Desktop\Tumor Detection\brain-tumor.mp4'

# Define the output video file path
video_path_out = '{}_out.mp4'.format(video_path)  # Output video file will be saved as input_out.mp4

# Open the video file
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()  # Read the first frame
H, W, _ = frame.shape  # Get video frame dimensions
fps = int(cap.get(cv2.CAP_PROP_FPS))  # Get video frame rate

# Setup output video writer
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), fps, (W, H))

# Load your custom-trained YOLO model (Tumor Detection)
model_path = r'C:\Users\VEDANT\OneDrive\Desktop\Tumor Detection\runs\detect\train3\weights\last.pt'
model = YOLO(model_path)  # Load your pre-trained model

# Detection threshold
threshold = 0.5

# Process each frame
while ret:
    # Run YOLO model to get detections
    results = model(frame)[0]

    # Draw bounding boxes and class names on the frame
    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            # Draw rectangle around the detected object
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            # Put the class label text
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    # Write processed frame to the output video
    out.write(frame)

    # Read the next frame
    ret, frame = cap.read()

# Release the video reader and writer resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Processed video saved at: {video_path_out}")
```

# building the output screenshot to new folder 


```python
import os
from ultralytics import YOLO
import cv2

# Define the input video file path
video_path = r'C:\Users\VEDANT\OneDrive\Desktop\Tumor Detection\input.mp4'

# Define the directory to save output
output_dir = r'C:\Users\VEDANT\OneDrive\Desktop\Tumor Detection\output'

# Function to create directory if it doesn't exist
def create_output_folder(folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Created folder: {folder_path}")
    else:
        print(f"Folder already exists: {folder_path}")

# Create the output directory
create_output_folder(output_dir)

# Define the output video file path
output_video_name = 'detected_output.mp4'
video_path_out = os.path.join(output_dir, output_video_name)  # Output video will be saved in the new folder

# Open the video file
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()  # Read the first frame
H, W, _ = frame.shape  # Get video frame dimensions
fps = int(cap.get(cv2.CAP_PROP_FPS))  # Get video frame rate

# Setup output video writer
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), fps, (W, H))

# Load your custom-trained YOLO model (Tumor Detection)
model_path = r'C:\Users\VEDANT\OneDrive\Desktop\Tumor Detection\runs\detect\train3\weights\best.pt'
model = YOLO(model_path)  # Load your pre-trained model

# Detection threshold
threshold = 0.5

# Variable to track if any object was detected
object_detected = False

# Counter for saving object images
image_counter = 0

# Process each frame
while ret:
    # Run YOLO model to get detections
    results = model(frame)[0]

    # Draw bounding boxes and class names on the frame
    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            # Mark that an object was detected
            object_detected = True

            # Draw rectangle around the detected object
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            # Put the class label text
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

            # Crop the detected object and save it as an image
            object_image = frame[int(y1):int(y2), int(x1):int(x2)]
            object_image_path = os.path.join(output_dir, f'detected_object_{image_counter}.jpg')
            cv2.imwrite(object_image_path, object_image)
            image_counter += 1

    # Write processed frame to the output video if any object was detected
    if object_detected:
        out.write(frame)

    # Read the next frame
    ret, frame = cap.read()

# Release the video reader and writer resources
cap.release()
out.release()
cv2.destroyAllWindows()

# Notify if the video was saved
if object_detected:
    print(f"Processed video saved at: {video_path_out}")
    print(f"Saved {image_counter} object images in the folder: {output_dir}")
else:
    print("No objects were detected, no video saved.")
```

```python
import os
from ultralytics import YOLO
import cv2

# Define the input image file path
image_path = r'test.jpg'

# Get the directory where the input image is located
base_dir = os.path.dirname(image_path)

# Load the input image
image = cv2.imread(image_path)

# Load your custom-trained YOLO model (Tumor Detection)
model_path = r'C:\Users\VEDANT\OneDrive\Desktop\Tumor Detection\runs\detect\train3\weights\best.pt'
model = YOLO(model_path)  # Load your pre-trained model

# Detection threshold
threshold = 0.5

# Counter for saving object images
image_counter = 0

# Run YOLO model to get detections
results = model(image)[0]

# Variable to track if any object was detected
object_detected = False

# Draw bounding boxes and class names on the image
for result in results.boxes.data.tolist():
    x1, y1, x2, y2, score, class_id = result

    if score > threshold:
        # Mark that an object was detected
        object_detected = True

        # Draw rectangle around the detected object
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
        # Put the class label text
        cv2.putText(image, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

        # Crop the detected object and save it as an image
        object_image = image[int(y1):int(y2), int(x1):int(x2)]
        object_image_path = os.path.join(base_dir, f'detected_object_{image_counter}.jpg')
        cv2.imwrite(object_image_path, object_image)
        image_counter += 1

# Save the processed image with bounding boxes
output_image_path = os.path.join(base_dir, 'processed_image_with_boxes.jpg')
cv2.imwrite(output_image_path, image)

# Notify if the image and detected objects were saved
if object_detected:
    print(f"Processed image saved at: {output_image_path}")
    print(f"Saved {image_counter} object images in the folder: {base_dir}")
else:
    print("No objects were detected.")
```
